```python
f = open('K07-netflix_titles_filtered.csv' ,encoding = 'ANSI')
reader = csv.reader(f)
next(reader)
```

```python
onto = get_ontology('ProjectOntology.owl').load()
iri = Namespace(onto.base_iri)
iri
```

```python
g = Graph()
g.bind("iri", iri)
g.bind("foaf", FOAF)
```

```python
for row in reader:
    try:
        type, title, director, cast, country, release_year, rating, duration, listed_in,desc = row 
        # print(type, title, director, cast, country, release_year, rating, duration, listed_in,desc)
        title =  Literal(title, datatype=XSD.string)
        cast = cast.split(', ')
        for i in range(len(cast)):
            cast[i] = Literal(cast[i], datatype=XSD.string)
        director = Literal(director, datatype=XSD.string)
        country= Literal(country, datatype=XSD.string)
        release_year = Literal(release_year, datatype=XSD.integer)
        rating = Literal(rating, datatype=XSD.string)
        duration = Literal(rating, datatype=XSD.string)
        listed_in = listed_in.split(', ')
        for i in range(len(listed_in)):
            listed_in[i] = Literal(listed_in[i], datatype=XSD.string)
        desc = Literal(desc, datatype=XSD.string)
        
        new_title = title.replace(" ","")
        iri.n_movie = URIRef((iri.netflixMovie+"/"+new_title)) 
        iri.n_cast = []
        for i in range(len(cast)):
            iri.n_cast.append(URIRef((iri.casts+"/"+new_title+str(i))))

        iri.n_listed_in = []
        for i in range(len(listed_in)):
            iri.n_listed_in.append(URIRef((iri.listed+"/"+new_title+str(i))))

        iri.new_title = URIRef((iri.Title+"/"+new_title))
        iri.newDirector = URIRef((iri.Director+"/"+new_title))
        iri.newDuration = URIRef((iri.Duration+"/"+new_title))
        iri.newRating = URIRef((iri.Rating+"/"+new_title))
        iri.newCountry = URIRef((iri.Country+"/"+new_title))
        iri.newRealeased_Year = URIRef((iri.Release_year+"/"+new_title))
        iri.newDesc = URIRef((iri.Desc+"/"+new_title))

        g.add((iri.n_movie,iri.hasTitle,iri.new_title))

        for i in range(len(iri.n_cast)):
            g.add((iri.n_movie,iri.hasCast,iri.n_cast[i]))
        
        for i in range(len(iri.n_listed_in)):
            g.add((iri.n_movie,iri.hasListedin,iri.n_listed_in[i]))

        g.add((iri.n_movie,iri.hasDirector,iri.newDirector))
        g.add((iri.n_movie,iri.hasDuration,iri.newDuration))
        g.add((iri.n_movie,iri.hasRating,iri.newRating))
        g.add((iri.n_movie,iri.hasCountry,iri.newCountry))
        g.add((iri.n_movie,iri.hasReleasedin,iri.newReleased_Year))
        g.add((iri.n_movie,iri.hasReleasedYear,iri.newReleased_Year))


    
        g.add((iri.new_title,RDFS.label,title))
        for i in range(len(iri.n_cast)):
            g.add((iri.n_cast[i],iri.isCastof,cast[i]))
        for i in range(len(iri.n_listed_in)):
            g.add((iri.n_listed_in[i],iri.isListedin,listed_in[i]))

        g.add((iri.newDirector,RDFS.label,director))

        g.add((iri.newDuration,RDFS.label,duration))

        g.add((iri.Rating,RDFS.label,rating))

        g.add((iri.newDesc,RDFS.label,desc))

        g.add((iri.Release_Year,RDFS.label,release_year))  
        g.add((iri.Country,RDFS.label,country))
    except:
        print('Error found')
    
```

```python
with open('K07-sparql.ttl', 'w', encoding='ANSI') as f:
  # do stuff ...
  f.write(g.serialize(format='turtle'))
  # f.close()
```
